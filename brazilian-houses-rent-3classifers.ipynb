```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
```

```python
data = pd.read_csv('../input/brasilian-houses-to-rent/houses_to_rent.csv')
```

```python
data.head(10)
```

```python
print(data.shape)
```

```python
data.info()
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data['floor'].sample(5)
```

```python
data['floor'].replace(to_replace = '-', value=0,inplace=True)
```

```python
data['floor'].head(10)
```

```python
data['animal'].replace(to_replace = 'not acept',value=0,inplace=True)
data['animal'].replace(to_replace = 'acept',value=1,inplace=True)
```

```python
data.sample(10)
```

```python
data['furniture'].replace(to_replace = 'not furnished',value=0,inplace=True)
data['furniture'].replace(to_replace = 'furnished',value=1,inplace=True)
```

```python
data.sample(5)
```

```python
data['rent amount'] = data['rent amount'].apply(lambda x: x.split('$')[1])
```

```python
data['rent amount'].head()
```

```python


data['fire insurance'] = data['fire insurance'].apply(lambda x: x.split('$')[1])
data['total'] = data['total'].apply(lambda x: x.split('$')[1])
```

```python
data.head()
```

```python
for col in ['hoa','property tax']:
    data[col].replace(to_replace='R\$', value='',regex=True, inplace=True)
    data[col].replace(to_replace=',', value='',regex=True, inplace=True)
```

```python
data.head()
```

```python
for col in ['rent amount','total']:
     data[col].replace(to_replace=',', value='',regex=True, inplace=True)
```

```python
data.head()
```

```python
data.describe()
```

```python
data['hoa'].replace(to_replace = 'Sem info',value=0,inplace=True)
data['hoa'].replace(to_replace = 'Incluso',value=0,inplace=True)
data['property tax'].replace(to_replace = 'Incluso',value=0,inplace=True)
```

```python
data.isin(['Incluso']).any()
```

```python
data = data.astype(dtype=np.int64)
```

```python
data.info()
```

```python
data = data.sample(frac=1).reset_index(drop=True)
```

```python
data.head()
```

```python
y = data['city']
X = data.drop('city',axis=1)
```

```python
X
```

```python
y
```

```python
scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```python
pd.DataFrame(X)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8)
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from sklearn import model_selection
from sklearn.utils import class_weight
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
log_model = LogisticRegression(penalty='l2')
svm_model = SVC()
nn_model = MLPClassifier(hidden_layer_sizes = (16,16),verbose=1,activation='relu', solver='adam')
```

```python
log_model.fit(X_train,y_train)
svm_model.fit(X_train,y_train)
nn_model.fit(X_train,y_train)
```

```python
print(log_model.score(X_test,y_test))
print(svm_model.score(X_test,y_test))
print(nn_model.score(X_test,y_test))
```
